| êµ¬ë¶„ | íšŒê·€ (Regression) | ë¶„ë¥˜ (Classification) |
|------|-------------------|----------------------|
| ì¶œë ¥ | ì—°ì†ê°’ (ì£¼ê°€) | ë²”ì£¼ (ìƒìŠ¹/í•˜ë½) |
| ì§ˆë¬¸ | ë‚´ì¼ ì£¼ê°€ëŠ” ì–¼ë§ˆ? | ë‚´ì¼ ì˜¤ë¥¼ê¹Œ ë‚´ë¦´ê¹Œ? |
| ì•Œê³ ë¦¬ì¦˜ | LinearRegression | LogisticRegression |
| í‰ê°€ | MSE, R2 | Accuracy, F1 |

### ì´ì§„ë¶„ë¥˜ ë¬¸ì œ ì •ì˜
```
ì…ë ¥(X): ì˜¤ëŠ˜ì˜ ì£¼ê°€ ê´€ë ¨ íŠ¹ì„±ë“¤ + í™˜ìœ¨ ë°ì´í„°
ì¶œë ¥(y): ë‚´ì¼ ìƒìŠ¹(1) ë˜ëŠ” í•˜ë½(0)
```

### LogisticRegression ì›ë¦¬
- ì´ë¦„ì— "Regression"ì´ ìˆì§€ë§Œ **ë¶„ë¥˜** ì•Œê³ ë¦¬ì¦˜
- Sigmoid í•¨ìˆ˜ë¡œ ì¶œë ¥ì„ 0~1 ì‚¬ì´ì˜ í™•ë¥ ë¡œ ë³€í™˜
- í™•ë¥ ì´ 0.5 ì´ìƒì´ë©´ 1(ìƒìŠ¹), ë¯¸ë§Œì´ë©´ 0(í•˜ë½)

$$
P(y=1 \mid X) = \frac{1}{1 + e^{-z}}
$$

$$
z = \beta_0 + \beta_1 x_1 + \cdots + \beta_n x_n
$$

## 3. íŠ¹ì„± ê³µí•™ (Feature Engineering)

ì£¼ê°€ ì˜ˆì¸¡ì— ìœ ìš©í•œ íŠ¹ì„±ë“¤ì„ ìƒì„±í•©ë‹ˆë‹¤. (20ì°¨ì‹œ ë‚´ìš© + í™˜ìœ¨ íŠ¹ì„± ì¶”ê°€)

````python
# íŠ¹ì„± ìƒì„± (í•µì‹¬ íŠ¹ì„±ë§Œ ê°„ë‹¨íˆ - 20ì°¨ì‹œì™€ ë™ì¼)
df['ì „ì¼ì¢…ê°€'] = df['ì¢…ê°€'].shift(1)           # ì „ì¼ ì¢…ê°€
df['ìˆ˜ìµë¥ '] = df['ì¢…ê°€'].pct_change() * 100   # ì „ì¼ ëŒ€ë¹„ ìˆ˜ìµë¥  (%)
df['5ì¼ì´ë™í‰ê· '] = df['ì¢…ê°€'].rolling(5).mean()  # 5ì¼ ì´ë™í‰ê· 
df['ê±°ë˜ëŸ‰ë¹„ìœ¨'] = df['ê±°ë˜ëŸ‰'] / df['ê±°ë˜ëŸ‰'].rolling(20).mean()  # í‰ê·  ëŒ€ë¹„ ê±°ë˜ëŸ‰

# í™˜ìœ¨ íŠ¹ì„± (FRED ë°ì´í„° í™œìš©)
if fx_available:
    df['í™˜ìœ¨ë³€í™”ìœ¨'] = df['í™˜ìœ¨'].pct_change() * 100  # ì „ì¼ ëŒ€ë¹„ í™˜ìœ¨ ë³€í™”ìœ¨ (%)

print("[ìƒì„±ëœ íŠ¹ì„±]")
print("=" * 50)
stock_features = ['ì „ì¼ì¢…ê°€', 'ìˆ˜ìµë¥ ', '5ì¼ì´ë™í‰ê· ', 'ê±°ë˜ëŸ‰ë¹„ìœ¨']
for feat in stock_features:
    print(f"- {feat}")

if fx_available:
    print(f"- í™˜ìœ¨ë³€í™”ìœ¨ (FRED)")
    print("\n* í™˜ìœ¨ì€ ì£¼ê°€ì™€ ìƒê´€ê´€ê³„ê°€ ìˆì„ ìˆ˜ ìˆìŒ")
    print("  (ì›í™” ì•½ì„¸ â†’ ìˆ˜ì¶œì£¼ ìƒìŠ¹ ê°€ëŠ¥ì„±)")
````

````python
# ê²°ì¸¡ì¹˜ ì œê±°
df = df.dropna()
print(f"\nê²°ì¸¡ì¹˜ ì œê±° í›„ ë°ì´í„°: {len(df)}ê°œ")
df.head()
````

````python
# íƒ€ê²Ÿ ë³€ìˆ˜ ìƒì„±: ë‹¤ìŒë‚  ìƒìŠ¹(1) / í•˜ë½(0)
df['ë‹¤ìŒë‚ ì¢…ê°€'] = df['ì¢…ê°€'].shift(-1)
df['ë‹¤ìŒë‚ ìƒìŠ¹'] = (df['ë‹¤ìŒë‚ ì¢…ê°€'] > df['ì¢…ê°€']).astype(int)
df = df.dropna()

print("[íƒ€ê²Ÿ ë³€ìˆ˜ ìƒì„±]")
print("=" * 50)
print("ë‹¤ìŒë‚ ìƒìŠ¹ = 1 if ë‹¤ìŒë‚ ì¢…ê°€ > ì˜¤ëŠ˜ì¢…ê°€ else 0")
print()
print(f"ìƒìŠ¹ ì¼ìˆ˜: {df['ë‹¤ìŒë‚ ìƒìŠ¹'].sum()}ì¼ ({df['ë‹¤ìŒë‚ ìƒìŠ¹'].mean()*100:.1f}%)")
print(f"í•˜ë½ ì¼ìˆ˜: {len(df) - df['ë‹¤ìŒë‚ ìƒìŠ¹'].sum()}ì¼ ({(1-df['ë‹¤ìŒë‚ ìƒìŠ¹'].mean())*100:.1f}%)")
````

````python
# ìƒìŠ¹/í•˜ë½ ë¶„í¬ ì‹œê°í™”
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# 1. íŒŒì´ ì°¨íŠ¸
labels = ['í•˜ë½ (0)', 'ìƒìŠ¹ (1)']
sizes = [len(df) - df['ë‹¤ìŒë‚ ìƒìŠ¹'].sum(), df['ë‹¤ìŒë‚ ìƒìŠ¹'].sum()]
colors = ['#ff6b6b', '#51cf66']
axes[0].pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
axes[0].set_title('íƒ€ê²Ÿ ë¶„í¬: ë‹¤ìŒë‚  ìƒìŠ¹/í•˜ë½')

# 2. ìˆ˜ìµë¥  íˆìŠ¤í† ê·¸ë¨
df['ë‹¤ìŒë‚ ìˆ˜ìµë¥ '] = (df['ë‹¤ìŒë‚ ì¢…ê°€'] - df['ì¢…ê°€']) / df['ì¢…ê°€'] * 100
axes[1].hist(df['ë‹¤ìŒë‚ ìˆ˜ìµë¥ '], bins=30, edgecolor='black', alpha=0.7, color='steelblue')
axes[1].axvline(0, color='red', linestyle='--', linewidth=2, label='0% (ê²½ê³„)')
axes[1].set_xlabel('ë‹¤ìŒë‚  ìˆ˜ìµë¥  (%)')
axes[1].set_ylabel('ë¹ˆë„')
axes[1].set_title('ë‹¤ìŒë‚  ìˆ˜ìµë¥  ë¶„í¬')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

````python
# í™˜ìœ¨ê³¼ ì£¼ê°€ì˜ ê´€ê³„ ì‹œê°í™”
if fx_available:
    fig, axes = plt.subplots(1, 2, figsize=(14, 4))
    
    # 1. í™˜ìœ¨ê³¼ ì¢…ê°€ ì¶”ì´
    ax1 = axes[0]
    ax2 = ax1.twinx()
    ax1.plot(df.index, df['ì¢…ê°€'], 'b-', label='ì‚¼ì„±ì „ì ì¢…ê°€', linewidth=1.5)
    ax2.plot(df.index, df['í™˜ìœ¨'], 'r-', label='USD/KRW í™˜ìœ¨', linewidth=1.5, alpha=0.7)
    ax1.set_xlabel('ë‚ ì§œ')
    ax1.set_ylabel('ì¢…ê°€ (ì›)', color='blue')
    ax2.set_ylabel('í™˜ìœ¨ (ì›)', color='red')
    ax1.tick_params(axis='x', rotation=45)
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')
    ax1.set_title('ì‚¼ì„±ì „ì ì¢…ê°€ vs USD/KRW í™˜ìœ¨')
    ax1.grid(True, alpha=0.3)
    
    # 2. í™˜ìœ¨ë³€í™”ìœ¨ê³¼ ìˆ˜ìµë¥  ì‚°ì ë„
    axes[1].scatter(df['í™˜ìœ¨ë³€í™”ìœ¨'], df['ìˆ˜ìµë¥ '], alpha=0.5, s=20)
    axes[1].axhline(0, color='gray', linestyle='--', linewidth=1)
    axes[1].axvline(0, color='gray', linestyle='--', linewidth=1)
    axes[1].set_xlabel('í™˜ìœ¨ ë³€í™”ìœ¨ (%)')
    axes[1].set_ylabel('ì£¼ê°€ ìˆ˜ìµë¥  (%)')
    axes[1].set_title('í™˜ìœ¨ ë³€í™”ìœ¨ vs ì£¼ê°€ ìˆ˜ìµë¥ ')
    axes[1].grid(True, alpha=0.3)
    
    # ìƒê´€ê³„ìˆ˜ í‘œì‹œ
    corr = df['í™˜ìœ¨ë³€í™”ìœ¨'].corr(df['ìˆ˜ìµë¥ '])
    axes[1].annotate(f'ìƒê´€ê³„ìˆ˜: {corr:.3f}', xy=(0.05, 0.95), xycoords='axes fraction',
                     fontsize=12, verticalalignment='top',
                     bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    
    plt.tight_layout()
    plt.show()
````

## 4. LogisticRegression ëª¨ë¸ í•™ìŠµ

### ì˜ˆì¸¡ ëŒ€ìƒ ì •ì˜
**íƒ€ê²Ÿ(y)**: ë‹¤ìŒë‚  ìƒìŠ¹(1) ë˜ëŠ” í•˜ë½(0)

````python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# íŠ¹ì„±ê³¼ íƒ€ê²Ÿ ë¶„ë¦¬ (í™˜ìœ¨ íŠ¹ì„± í¬í•¨)
if fx_available:
    feature_cols = ['ì „ì¼ì¢…ê°€', 'ìˆ˜ìµë¥ ', '5ì¼ì´ë™í‰ê· ', 'ê±°ë˜ëŸ‰ë¹„ìœ¨', 'í™˜ìœ¨ë³€í™”ìœ¨']
else:
    feature_cols = ['ì „ì¼ì¢…ê°€', 'ìˆ˜ìµë¥ ', '5ì¼ì´ë™í‰ê· ', 'ê±°ë˜ëŸ‰ë¹„ìœ¨']

X = df[feature_cols]
y = df['ë‹¤ìŒë‚ ìƒìŠ¹']

print("[ë°ì´í„°ì…‹ êµ¬ì„±]")
print("=" * 50)
print(f"íŠ¹ì„± (X): {len(feature_cols)}ê°œ")
for col in feature_cols:
    print(f"  - {col}")
print(f"\níƒ€ê²Ÿ (y): ë‹¤ìŒë‚ ìƒìŠ¹ (0=í•˜ë½, 1=ìƒìŠ¹)")
print(f"ìƒ˜í”Œ ìˆ˜: {len(X)}ê°œ")
````

````python
# ë°ì´í„° ë¶„í•  (ì‹œê³„ì—´ì´ë¯€ë¡œ ì„ì§€ ì•ŠìŒ - 80:20)
split_idx = int(len(X) * 0.8)

X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]

print(f"\n[ë°ì´í„° ë¶„í• ]")
print(f"í•™ìŠµ ë°ì´í„°: {len(X_train)}ê°œ")
print(f"í…ŒìŠ¤íŠ¸ ë°ì´í„°: {len(X_test)}ê°œ")
print(f"í…ŒìŠ¤íŠ¸ ìƒìŠ¹ ë¹„ìœ¨: {y_test.mean():.1%}")
````

````python
# ìŠ¤ì¼€ì¼ë§ (StandardScaler)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("[ìŠ¤ì¼€ì¼ë§ ì™„ë£Œ]")
print(f"í•™ìŠµ ë°ì´í„° í‰ê· : {X_train_scaled.mean():.4f}")
print(f"í•™ìŠµ ë°ì´í„° í‘œì¤€í¸ì°¨: {X_train_scaled.std():.4f}")
````

````python
# LogisticRegression ëª¨ë¸ í•™ìŠµ
model = LogisticRegression(random_state=42, max_iter=1000)
model.fit(X_train_scaled, y_train)

print("[LogisticRegression ëª¨ë¸ í•™ìŠµ ì™„ë£Œ]")
````

````python
# ì˜ˆì¸¡
y_pred = model.predict(X_test_scaled)
y_pred_proba = model.predict_proba(X_test_scaled)[:, 1]  # ìƒìŠ¹ í™•ë¥ 
````

---
## 5. ëª¨ë¸ í‰ê°€ ë° ì‹œê°í™”

### ë¶„ë¥˜ í‰ê°€ ì§€í‘œ
| ì§€í‘œ | ìˆ˜ì‹ | ì˜ë¯¸ |
|------|------|------|
| Accuracy | (TP+TN)/ì „ì²´ | ì „ì²´ ì •í™•ë„ |
| Precision | TP/(TP+FP) | ìƒìŠ¹ ì˜ˆì¸¡ ì¤‘ ì‹¤ì œ ìƒìŠ¹ |
| Recall | TP/(TP+FN) | ì‹¤ì œ ìƒìŠ¹ ì¤‘ ì˜ˆì¸¡ ìƒìŠ¹ |
| F1 Score | 2*(P*R)/(P+R) | Precisionê³¼ Recall ì¡°í™”í‰ê·  |

- TP: True Positive (ìƒìŠ¹ ì˜ˆì¸¡ â†’ ì‹¤ì œ ìƒìŠ¹)
- TN: True Negative (í•˜ë½ ì˜ˆì¸¡ â†’ ì‹¤ì œ í•˜ë½)
- FP: False Positive (ìƒìŠ¹ ì˜ˆì¸¡ â†’ ì‹¤ì œ í•˜ë½)
- FN: False Negative (í•˜ë½ ì˜ˆì¸¡ â†’ ì‹¤ì œ ìƒìŠ¹)

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# í‰ê°€ ì§€í‘œ ê³„ì‚°
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, zero_division=0)
recall = recall_score(y_test, y_pred, zero_division=0)
f1 = f1_score(y_test, y_pred, zero_division=0)

print("[ëª¨ë¸ í‰ê°€ ê²°ê³¼]")
print("=" * 50)
print(f"Accuracy:  {accuracy:.1%} (ì „ì²´ ì¤‘ ë§ì¶˜ ë¹„ìœ¨)")
print(f"Precision: {precision:.1%} (ìƒìŠ¹ ì˜ˆì¸¡ ì¤‘ ì‹¤ì œ ìƒìŠ¹)")
print(f"Recall:    {recall:.1%} (ì‹¤ì œ ìƒìŠ¹ ì¤‘ ì˜ˆì¸¡ ìƒìŠ¹)")
print(f"F1 Score:  {f1:.1%} (Precisionê³¼ Recall ì¡°í™”í‰ê· )")
````

````python
# Classification Report
print("\n[Classification Report]")
print("=" * 50)
print(classification_report(y_test, y_pred, target_names=['í•˜ë½', 'ìƒìŠ¹'], zero_division=0))
````
