## 2. ë°ì´í„° ìˆ˜ì§‘ (ì£¼ê°€ + í™˜ìœ¨)

### 2.1 ì£¼ê°€ ë°ì´í„° ìˆ˜ì§‘ (FinanceDataReader)

````python
import FinanceDataReader as fdr

# ë‚ ì§œ ì„¤ì • (ìµœê·¼ 1ë…„)
end_date = datetime.now()
start_date = end_date - timedelta(days=365)

print("[ì£¼ê°€ ë°ì´í„° ìˆ˜ì§‘]")
print("=" * 50)
print(f"ê¸°ê°„: {start_date.strftime('%Y-%m-%d')} ~ {end_date.strftime('%Y-%m-%d')}")
print("ì¢…ëª©: ì‚¼ì„±ì „ì (005930)")

# ì‚¼ì„±ì „ì ì£¼ê°€ ë°ì´í„° ìˆ˜ì§‘ (FinanceDataReader - Module 01/02ì™€ ë™ì¼)
df = fdr.DataReader("005930", start_date, end_date)

print(f"\nìˆ˜ì§‘ëœ ë°ì´í„°: {len(df)}ê°œ")
display(df.tail())
````

### 2.2 í™˜ìœ¨ ë°ì´í„° ìˆ˜ì§‘ (FRED API)

FREDì—ì„œ USD/KRW í™˜ìœ¨ ë°ì´í„°ë¥¼ ê°€ì ¸ì˜µë‹ˆë‹¤. (Module 02ì—ì„œ í•™ìŠµí•œ ë‚´ìš© í™œìš©)

````python
print("[í™˜ìœ¨ ë°ì´í„° ìˆ˜ì§‘ - FRED]")
print("=" * 50)
print("ì‹œë¦¬ì¦ˆ: DEXKOUS (USD/KRW í™˜ìœ¨)")

# FREDì—ì„œ í™˜ìœ¨ ë°ì´í„° ìˆ˜ì§‘
df_fx = web.DataReader('DEXKOUS', 'fred', start_date, end_date)
df_fx.columns = ['í™˜ìœ¨']

# ê²°ì¸¡ì¹˜ ì²˜ë¦¬ (ì£¼ë§/ê³µíœ´ì¼ â†’ ì „ì¼ ê°’ìœ¼ë¡œ ì±„ì›€)
df_fx = df_fx.ffill()

df_fx.tail()
````

````python
# ì£¼ê°€ ë°ì´í„°ì™€ í™˜ìœ¨ ë°ì´í„° ë³‘í•©
df = df.merge(df_fx, left_index=True, right_index=True, how='left')
df.head()
````

````python
# í™˜ìœ¨ ê²°ì¸¡ì¹˜ ì²˜ë¦¬ (ì£¼ì‹ ê±°ë˜ì¼ê³¼ í™˜ìœ¨ ë°ì´í„° ë¶ˆì¼ì¹˜ ì‹œ)
df['í™˜ìœ¨'] = df['í™˜ìœ¨'].ffill().bfill()  # ì•ì˜ ê°’ìœ¼ë¡œ ë¨¼ì € ì±„ìš°ê³ , ê·¸ë˜ë„ ë‚¨ìœ¼ë©´ ë’¤ì˜ ê°’ìœ¼ë¡œ ì±„ìš´ë‹¤
````

## 3. íŠ¹ì„± ê³µí•™ (Feature Engineering)

ì£¼ê°€ ì˜ˆì¸¡ì— ìœ ìš©í•œ íŠ¹ì„±ë“¤ì„ ìƒì„±í•©ë‹ˆë‹¤. (20ì°¨ì‹œ ë‚´ìš© + í™˜ìœ¨ íŠ¹ì„± ì¶”ê°€)

````python
# íŠ¹ì„± ìƒì„± (í•µì‹¬ íŠ¹ì„±ë§Œ ê°„ë‹¨íˆ - 20ì°¨ì‹œì™€ ë™ì¼)
df['ì „ì¼ì¢…ê°€'] = df['Close'].shift(1)           # ì „ì¼ ì¢…ê°€
df['ìˆ˜ìµë¥ '] = df['Close'].pct_change() * 100   # ì „ì¼ ëŒ€ë¹„ ìˆ˜ìµë¥  (%)
df['5ì¼ì´ë™í‰ê· '] = df['Close'].rolling(5).mean()  # 5ì¼ ì´ë™í‰ê· 
df['ê±°ë˜ëŸ‰ë¹„ìœ¨'] = df['Volume'] / df['Volume'].rolling(20).mean()  # í‰ê·  ëŒ€ë¹„ ê±°ë˜ëŸ‰

# í™˜ìœ¨ íŠ¹ì„± (FRED ë°ì´í„° í™œìš©)
df['í™˜ìœ¨ë³€í™”ìœ¨'] = df['í™˜ìœ¨'].pct_change() * 100  # ì „ì¼ ëŒ€ë¹„ í™˜ìœ¨ ë³€í™”ìœ¨ (%)

# ê²°ì¸¡ì¹˜ ì œê±°
df = df.dropna()
df.head()
````

````python
# íƒ€ê²Ÿ ë³€ìˆ˜ ìƒì„±: ë‹¤ìŒë‚  ìƒìŠ¹(1) / í•˜ë½(0)
df['ë‹¤ìŒë‚ ì¢…ê°€'] = df['Close'].shift(-1)
df['ë‹¤ìŒë‚ ìƒìŠ¹'] = (df['ë‹¤ìŒë‚ ì¢…ê°€'] > df['Close']).astype(int)
df = df.dropna()
df.head()
````

---
## 4. LogisticRegression ëª¨ë¸ í•™ìŠµ

### ì˜ˆì¸¡ ëŒ€ìƒ ì •ì˜
**íƒ€ê²Ÿ(y)**: ë‹¤ìŒë‚  ìƒìŠ¹(1) ë˜ëŠ” í•˜ë½(0)

````python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# íŠ¹ì„±ê³¼ íƒ€ê²Ÿ ë¶„ë¦¬ (í™˜ìœ¨ íŠ¹ì„± í¬í•¨)
feature_cols = ['ì „ì¼ì¢…ê°€', 'ìˆ˜ìµë¥ ', '5ì¼ì´ë™í‰ê· ', 'ê±°ë˜ëŸ‰ë¹„ìœ¨', 'í™˜ìœ¨ë³€í™”ìœ¨']

X = df[feature_cols]
y = df['ë‹¤ìŒë‚ ìƒìŠ¹']
X.shape, y.shape
````

````python
# ë°ì´í„° ë¶„í•  (ì‹œê³„ì—´ì´ë¯€ë¡œ ì„ì§€ ì•ŠìŒ - 80:20)
split_idx = int(len(X) * 0.8)

X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]

X_train.shape, X_test.shape, y_train.shape, y_test.shape
````

````python
# ìŠ¤ì¼€ì¼ë§ (StandardScaler)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
````

````python
# LogisticRegression ëª¨ë¸ í•™ìŠµ
model = LogisticRegression(random_state=42, max_iter=1000)
model.fit(X_train_scaled, y_train)
````

````python
# ì˜ˆì¸¡
y_pred = model.predict(X_test_scaled)
y_pred_proba = model.predict_proba(X_test_scaled)[:, 1]  # ìƒìŠ¹ í™•ë¥ 
````

````python
from sklearn.metrics import accuracy_score

# í‰ê°€ ì§€í‘œ ê³„ì‚°
accuracy = accuracy_score(y_test, y_pred)

print("[ëª¨ë¸ í‰ê°€ ê²°ê³¼]")
print("=" * 50)
print(f"Accuracy:  {accuracy:.1%} (ì „ì²´ ì¤‘ ë§ì¶˜ ë¹„ìœ¨)")
````
