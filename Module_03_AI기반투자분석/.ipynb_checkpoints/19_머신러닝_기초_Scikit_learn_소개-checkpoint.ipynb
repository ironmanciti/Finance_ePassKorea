## 1. 머신러닝이란?

### 정의
**머신러닝(Machine Learning)**: 명시적으로 프로그래밍하지 않아도 데이터로부터 패턴을 학습하여 예측하거나 결정을 내리는 알고리즘

### 전통적 프로그래밍 vs 머신러닝

| 구분 | 전통적 프로그래밍 | 머신러닝 |
|------|------------------|----------|
| 입력 | 데이터 + 규칙 | 데이터 + 정답 |
| 출력 | 결과 | 규칙(모델) |
| 예시 | if 주가 > 이동평균: 매수 | 과거 데이터로 매수 시점 학습 |

### 머신러닝의 3단계
```
1. 학습(Training): 과거 데이터로 패턴 학습
2. 검증(Validation): 모델 성능 평가
3. 예측(Prediction): 새로운 데이터에 적용
```

## 2. 지도학습 vs 비지도학습

### 2.1 지도학습 (Supervised Learning)

**정의**: 입력(X)과 정답(y)이 함께 주어진 데이터로 학습

| 유형 | 설명 | 금융 활용 예시 |
|------|------|----------------|
| **분류(Classification)** | 범주 예측 | 주가 상승/하락 예측, 부도 여부 판단 |
| **회귀(Regression)** | 연속값 예측 | 주가 예측, 수익률 예측 |

```
[지도학습 예시]
입력(X): 거래량, PER, PBR, 이동평균
정답(y): 다음날 상승(1) / 하락(0)
→ 모델이 X와 y의 관계를 학습
```

### 2.2 비지도학습 (Unsupervised Learning)

**정의**: 정답(y) 없이 입력(X)만으로 데이터의 구조나 패턴을 학습

| 유형 | 설명 | 금융 활용 예시 |
|------|------|----------------|
| **군집화(Clustering)** | 유사한 데이터 그룹화 | 고객 세분화, 종목 그룹화 |
| **차원축소(Dimensionality Reduction)** | 변수 압축 | 다수 지표를 핵심 요인으로 축소 |

```
[비지도학습 예시]
입력(X): 여러 종목의 수익률, 변동성, 거래량
→ 모델이 자동으로 유사한 종목끼리 그룹화
```

````python
# 지도학습 vs 비지도학습 시각적 비교
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# 지도학습 예시: 분류
np.random.seed(42)
X_up = np.random.randn(30, 2) + [2, 2]       # 상승 종목
X_down = np.random.randn(30, 2) + [-1, -1]  # 하락 종목

axes[0].scatter(X_up[:, 0], X_up[:, 1], c='green', label='상승', alpha=0.7, s=50)
axes[0].scatter(X_down[:, 0], X_down[:, 1], c='red', label='하락', alpha=0.7, s=50)
axes[0].axhline(y=0.5, color='black', linestyle='--', label='결정 경계')

axes[0].set_xlabel('특성 1 (예: 거래량 변화)')
axes[0].set_ylabel('특성 2 (예: 가격 모멘텀)')
axes[0].set_title('지도학습: 분류')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# 비지도학습 예시: 군집화
X_cluster = np.vstack([
    np.random.randn(20, 2) + [0, 3],
    np.random.randn(20, 2) + [3, 0],
    np.random.randn(20, 2) + [-2, -2]
])
colors = ['blue'] * 20 + ['orange'] * 20 + ['purple'] * 20

axes[1].scatter(X_cluster[:, 0], X_cluster[:, 1], c=colors, alpha=0.7, s=50)
axes[1].set_xlabel('특성 1')
axes[1].set_ylabel('특성 2')
axes[1].set_title('비지도학습: 군집화')
axes[1].grid(True, alpha=0.3)

# 군집 중심 표시
for center, color in zip([[0, 3], [3, 0], [-2, -2]], ['blue', 'orange', 'purple']):
    axes[1].scatter(*center, c=color, marker='X', s=200,
                    edgecolors='black', linewidths=2)

plt.tight_layout()
plt.show()
````

## 3. 금융 분야 활용 사례

### 지도학습 활용
| 문제 | 알고리즘 | 설명 |
|------|----------|------|
| 주가 방향 예측 | 로지스틱 회귀, 랜덤 포레스트 | 다음날 상승/하락 분류 |

| 이상 거래 탐지 | Isolation Forest | 비정상 패턴 탐지 |

## 4. Scikit-learn 기초

### Scikit-learn이란?
- Python의 대표적인 머신러닝 라이브러리
- 다양한 알고리즘을 **일관된 인터페이스**로 제공
- 공식 문서: https://scikit-learn.org

### 핵심 API 패턴
```python
from sklearn.모듈 import 알고리즘

# 1. 모델 생성
model = 알고리즘()
# 2. 학습 (지도학습)
model.fit(X_train, y_train)
# 3. 예측
y_pred = model.predict(X_test)
# 4. 평가
score = model.score(X_test, y_test)
```

````python
# Scikit-learn 기본 import
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, classification_report

print("[Scikit-learn 주요 모듈]")
print("=" * 50)
print("sklearn.model_selection : 데이터 분할, 교차검증")
print("sklearn.preprocessing   : 데이터 전처리 (스케일링)")
print("sklearn.linear_model    : 선형 모델 (회귀, 분류)")
print("sklearn.tree            : 의사결정나무")
print("sklearn.ensemble        : 앙상블 (랜덤포레스트)")
print("sklearn.cluster         : 군집화 (K-Means)")
print("sklearn.metrics         : 평가 지표")
````

## 5. 간단한 실습 예제

### 5.1 지도학습: 주가 방향 예측 (분류)

````python
# 가상의 주식 데이터 생성
np.random.seed(42)
n_samples = 200

# 특성: 거래량 변화율
volume_change = np.random.randn(n_samples)

# 타겟: 거래량이 일정 수준 이상이면 상승
target = (volume_change + np.random.randn(n_samples) * 0.5 > 0).astype(int)

df_stock = pd.DataFrame({
    '거래량변화율': volume_change,
    '다음날상승': target
})

print("[가상 주식 데이터]")
print("=" * 50)
display(df_stock.head(10))
print(f"\n데이터 크기: {len(df_stock)}개")
print(f"상승 비율: {df_stock['다음날상승'].mean():.1%}")
````

````python
# 데이터 분할: 학습용 / 테스트용
X = df_stock[['거래량변화율']]
y = df_stock['다음날상승']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print("[데이터 분할]")
print("=" * 50)
print(f"학습 데이터: {len(X_train)}개")
print(f"테스트 데이터: {len(X_test)}개")
````

````python
# 모델 학습: 로지스틱 회귀
model = LogisticRegression()
model.fit(X_train, y_train)

# 예측
y_pred = model.predict(X_test)

# 평가
accuracy = accuracy_score(y_test, y_pred)

print("[지도학습: 로지스틱 회귀 결과]")
print("=" * 50)
print(f"정확도(Accuracy): {accuracy:.1%}")
print("\n[분류 리포트]")
print(classification_report(y_test, y_pred, target_names=['하락', '상승']))
````

### 5.2 비지도학습: 종목 군집화

````python
# 가상의 종목 데이터 생성
np.random.seed(123)

# 3가지 유형의 종목 (성장주, 가치주, 배당주)
growth_stocks = np.random.randn(15, 2) + [3, 3]     # 고성장, 고변동성
value_stocks = np.random.randn(15, 2) + [-1, 1]     # 저평가, 중변동성
dividend_stocks = np.random.randn(15, 2) + [0, -2]  # 안정, 저변동성

X_stocks = np.vstack([growth_stocks, value_stocks, dividend_stocks])

df_cluster = pd.DataFrame(X_stocks, columns=['Return', 'Volatility'])
df_cluster['StockID'] = [f'Stock_{i+1}' for i in range(45)]

print("[가상 종목 데이터 (정답 없음)]")
print("=" * 50)
display(df_cluster.head())
````

````python
# K-Means 군집화
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df_cluster['Cluster'] = kmeans.fit_predict(X_stocks)

print("[비지도학습: K-Means 군집화 결과]")
print("=" * 50)
print(df_cluster.groupby('Cluster').agg({
    'Return': ['mean', 'std'],
    'Volatility': ['mean', 'std'],
    'StockID': 'count'
}).round(2))
````

````python
# 군집화 결과 시각화
fig, ax = plt.subplots(figsize=(8, 6))

# 색상 ↔ 의미 매칭 (고성장 → 저성장 → 안정)
colors = ['green', 'blue', 'orange']

for i in range(3):
    mask = df_cluster['Cluster'] == i
    ax.scatter(
        df_cluster.loc[mask, 'Return'],
        df_cluster.loc[mask, 'Volatility'],
        c=colors[i],
        alpha=0.7,
        s=80,
        edgecolors='black'
    )


# 🔹 한글화된 label / title
ax.set_xlabel('기대 수익률')
ax.set_ylabel('변동성')
ax.set_title('K-평균 군집화: 종목 유형 분류')

ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````
