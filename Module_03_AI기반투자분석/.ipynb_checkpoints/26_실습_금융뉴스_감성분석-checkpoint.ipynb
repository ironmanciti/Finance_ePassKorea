## 2. 금융 뉴스 수집 (모듈 2 연계)

````python
# 모듈 2에서 배운 크롤링 함수 (15차시)
HEADERS = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
}

def get_soup(url):
    """URL에서 BeautifulSoup 객체 반환"""
    response = requests.get(url, headers=HEADERS, timeout=10)
    response.raise_for_status()
    return BeautifulSoup(response.text, 'html.parser')

def crawl_financial_news(limit=20):
    """네이버 금융 주요 뉴스 크롤링"""
    url = "https://finance.naver.com/news/mainnews.naver"
    soup = get_soup(url)
    now = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

    news_data = []
    news_items = soup.select('ul.newsList li')[:limit]

    for item in news_items:
        title_tag = item.select_one('dd.articleSubject a')
        summary_tag = item.select_one('dd.articleSummary')

        if title_tag:
            title = title_tag.get_text(strip=True)
            link = 'https://finance.naver.com' + title_tag.get('href', '')

            summary = ''
            if summary_tag:
                summary = summary_tag.get_text(strip=True)

            news_data.append({
                '제목': title,
                '요약': summary[:100] if summary else '',
                '링크': link,
                '수집시각': now
            })

    return pd.DataFrame(news_data)
````

````python
# 뉴스 수집
print("[금융 뉴스 크롤링]")
print("=" * 60)

df_news = crawl_financial_news(limit=20)
print(f"수집된 뉴스: {len(df_news)}건")
display(df_news[['제목']].head(10))
````

## 3. 감성 사전 기반 분석

### 금융 감성 사전 정의
금융 도메인에 특화된 긍정/부정 단어 사전을 구축합니다.

````python
# 금융 감성 사전 정의
positive_words = [
    # 상승 관련
    '상승', '급등', '폭등', '오름', '올라', '상향', '반등', '회복',
    # 긍정 실적
    '호조', '호실적', '흑자', '최대', '사상최고', '신고가', '돌파',
    # 투자 긍정
    '매수', '순매수', '유입', '확대', '성장', '증가', '개선',
    # 전망 긍정
    '기대', '낙관', '호재', '수혜', '강세', '우상향', '목표가상향'
]

negative_words = [
    # 하락 관련
    '하락', '급락', '폭락', '떨어', '내려', '하향', '추락', '약세',
    # 부정 실적
    '부진', '악화', '적자', '최저', '사상최저', '저조', '감소',
    # 투자 부정
    '매도', '순매도', '유출', '축소', '위축', '감소', '악화',
    # 전망 부정
    '우려', '비관', '악재', '리스크', '위기', '불안', '경고',
    # 경제 부정
    '불황', '침체', '긴축', '인상'
]

print("[금융 감성 사전]")
print("=" * 60)
print(f"긍정 단어: {len(positive_words)}개")
print(f"예시: {positive_words[:10]}")
print(f"\n부정 단어: {len(negative_words)}개")
print(f"예시: {negative_words[:10]}")
````

````python
# 감성 분석 함수
def analyze_sentiment(text, positive_words, negative_words):
    """
    텍스트의 감성 점수 계산

    Returns:
        dict: pos_count, neg_count, score, label
    """
    # 긍정/부정 단어 카운트
    pos_count = sum(1 for word in positive_words if word in text)
    neg_count = sum(1 for word in negative_words if word in text)

    # 감성 점수 (-1 ~ 1)
    total = pos_count + neg_count
    if total == 0:
        score = 0
    else:
        score = (pos_count - neg_count) / total

    # 레이블
    if score > 0.1:
        label = '긍정'
    elif score < -0.1:
        label = '부정'
    else:
        label = '중립'

    return {
        'pos_count': pos_count,
        'neg_count': neg_count,
        'score': score,
        'label': label
    }

# 테스트
test_texts = [
    "삼성전자 3분기 영업이익 급등, 사상최고 실적 기대",
    "코스피 하락세 지속, 외국인 순매도 확대에 우려",
    "현대차 신차 출시, 시장 반응 주목"
]

print("[감성 분석 테스트]")
print("=" * 60)
for text in test_texts:
    result = analyze_sentiment(text, positive_words, negative_words)
    print(f"텍스트: {text}")
    print(f"  → 긍정: {result['pos_count']}, 부정: {result['neg_count']}")
    print(f"  → 점수: {result['score']:.2f}, 레이블: {result['label']}")
    print()
````

---
## 4. 뉴스 감성 점수 계산

````python
# 전체 뉴스에 감성 분석 적용
def analyze_news_sentiment(df, positive_words, negative_words):
    """뉴스 DataFrame에 감성 분석 결과 추가"""
    results = []

    for _, row in df.iterrows():
        # 제목 + 요약 합쳐서 분석
        text = row['제목'] + ' ' + str(row.get('요약', ''))
        result = analyze_sentiment(text, positive_words, negative_words)
        results.append(result)

    # 결과를 DataFrame에 추가
    df_result = df.copy()
    df_result['긍정단어수'] = [r['pos_count'] for r in results]
    df_result['부정단어수'] = [r['neg_count'] for r in results]
    df_result['감성점수'] = [r['score'] for r in results]
    df_result['감성레이블'] = [r['label'] for r in results]

    return df_result

# 감성 분석 적용
df_sentiment = analyze_news_sentiment(df_news, positive_words, negative_words)

print("[뉴스 감성 분석 결과]")
print("=" * 60)
display(df_sentiment[['제목', '긍정단어수', '부정단어수', '감성점수', '감성레이블']].head(10))
````

````python
# 감성 분포 통계
print("[감성 분포 통계]")
print("=" * 60)

sentiment_counts = df_sentiment['감성레이블'].value_counts()
print(sentiment_counts)

avg_score = df_sentiment['감성점수'].mean()
print(f"\n평균 감성 점수: {avg_score:.3f}")

if avg_score > 0.05:
    market_sentiment = "긍정적 (낙관적)"
elif avg_score < -0.05:
    market_sentiment = "부정적 (비관적)"
else:
    market_sentiment = "중립적"

print(f"시장 심리: {market_sentiment}")
````

## 5. 시각화 및 해석

````python
# 감성 분포 시각화
fig, axes = plt.subplots(1, 2, figsize=(15, 4))

# 1. 감성 레이블 분포
colors = {'긍정': 'green', '중립': 'gray', '부정': 'red'}
sentiment_counts = df_sentiment['감성레이블'].value_counts()
axes[0].bar(
    sentiment_counts.index,
    sentiment_counts.values,
    color=[colors.get(x, 'gray') for x in sentiment_counts.index],
    alpha=0.7,
    edgecolor='black'
)
axes[0].set_xlabel('감성')
axes[0].set_ylabel('개수')
axes[0].set_title('뉴스 감성 분포')
axes[0].grid(True, alpha=0.3, axis='y')

# 2. 파이 차트
axes[1].pie(
    sentiment_counts,
    labels=sentiment_counts.index,
    autopct='%1.1f%%',
    colors=[colors.get(x, 'gray') for x in sentiment_counts.index],
    startangle=90
)
axes[1].set_title('감성 비율')

plt.tight_layout()
plt.show()
````

````python
# 감성별 뉴스 예시
print("[감성별 뉴스 예시]")
print("=" * 60)

for label in ['긍정', '부정', '중립']:
    subset = df_sentiment[df_sentiment['감성레이블'] == label]
    if len(subset) > 0:
        print(f"\n[{label} 뉴스]")
        for _, row in subset.head(3).iterrows():
            print(f"  - {row['제목'][:50]}...")
            print(f"    점수: {row['감성점수']:.2f}")
````

## 감성 분석 결과 해석

````python
print("[시장 심리 종합 분석]")
print("=" * 60)

total_news = len(df_sentiment)
pos_ratio = (df_sentiment['감성레이블'] == '긍정').sum() / total_news * 100
neg_ratio = (df_sentiment['감성레이블'] == '부정').sum() / total_news * 100
neutral_ratio = (df_sentiment['감성레이블'] == '중립').sum() / total_news * 100

print(f"\n분석 뉴스 수: {total_news}건")
print(f"긍정 뉴스 비율: {pos_ratio:.1f}%")
print(f"부정 뉴스 비율: {neg_ratio:.1f}%")
print(f"중립 뉴스 비율: {neutral_ratio:.1f}%")
print(f"\n평균 감성 점수: {avg_score:.3f}")

print("\n[해석]")
if avg_score > 0.1:
    print("현재 금융 뉴스는 전반적으로 긍정적인 톤입니다.")
    print("시장 참여자들의 낙관적 심리가 반영될 수 있습니다.")
elif avg_score < -0.1:
    print("현재 금융 뉴스는 전반적으로 부정적인 톤입니다.")
    print("시장 참여자들의 비관적 심리가 반영될 수 있습니다.")
else:
    print("현재 금융 뉴스는 중립적인 톤입니다.")
    print("뚜렷한 시장 방향성 신호가 없습니다.")

print("\n주의: 감성 분석은 참고용이며 투자 의사결정의 유일한 근거가 되어서는 안 됩니다.")
````

### 2. 감성 사전 기반 분석
```python
def analyze_sentiment(text, pos_words, neg_words):
    pos_count = sum(1 for w in pos_words if w in text)
    neg_count = sum(1 for w in neg_words if w in text)
    
    total = pos_count + neg_count
    score = (pos_count - neg_count) / total if total > 0 else 0
    
    return score
```
